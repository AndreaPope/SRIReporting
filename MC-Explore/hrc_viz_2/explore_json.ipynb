```python
import pandas as pd
```

```python
df2 = pd.read_csv('data/SRI-R-HRC.csv')
```

```python
df2 = df2[df2['CEI_Score_2017'] == 100]
df2.head()
```

```python
df2['new_mkt'] = df2.MarketCap.apply(lambda x: float(x[1:-1]))
df2.head()
```

```python
df2 = df2[['Sector', 'Symbol', 'LastSale', 'new_mkt']]
df2.columns = ['name', 'Symbol', 'LastSale', 'new_mkt']
df2.dropna(inplace=True)
df2.head()
```

```python
len(df2)
```

```python
j = (df2.groupby(['name'], as_index=True)
        .apply(lambda x: x[['Symbol','new_mkt']].to_dict('r'))
        .reset_index()
        .rename(columns={0:'children'})
        .to_json('data/test_19NOV.json', orient='records'))
```
